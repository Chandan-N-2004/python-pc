**GRADE SYSTEM FOR SCORE**

```python
score = int(input("Score : "))
    
if score >= 90 and score <= 100:  # score >=90 :   #
        print("Grade A")
    
elif score >= 80 and score < 90:   # score >=80 :
        print("Grade B")
        
elif score >= 70 and score < 80:   # score >=70 :
        print("Grade C")    
        
else:
        print("FAIL \n Better luck next time")
```

***" MATCH " concepts***

```python

name = input ("enter ur name ? ")

if name == 'chandan' :
    print ('califorina')
elif name == 'james' :
    print ('texas')
elif name == 'john' :
    print ('newyork')
else :
    print ('who ??')
    
# above code can be written in the below way by using match concepts
```

```python
# by using the match concept we can write the above code 
     # supports only on( python 3.10 version)
name = input("enter ur name? ")

match name:
    case 'chandan':
        print('california')
# as we can print the same o/p for diiferent names by inserting "|" in between the names
    case 'suresh' | 'james' |'mathive' :
        print('south africa')
    case 'james':
        print('texas')
    case 'john':
        print('newyork')
    case _:
        print('who??')
```
